```python
## label,color,marker,grid ,axis,legend,text,title
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(11)
y=x**3
y2=x**2
print(y)
print(x)
plt.plot(x,y,'black',marker='o',label='x**3')            ## label it help to show legends with its line graph details
plt.plot(x,y2,'red',marker='o',linewidth=6,label='x**2')  ## overridden with specified linewidth 
plt.ylabel('y')                            ## specifying label for y axis
plt.xlabel('x')                               ## specifying label for x axis
plt.title('Matplotlib Demo')                  ## spcifying Graph title 
plt.legend()                                  ## specifying legend for each line plot
plt.text(5,125,'text=5**3',fontsize=20)       ## specifying text for coordinate
plt.axis([0,20,0,500])                        ## plt.axis ([x,y]), we can change our axis according to our data by this function
plt.grid()
plt.show()
```

## Bubble Chart
* used to show the size of other data points , for example - population size , for salary and years 

```python
salary=[100,150,200,220,330,100]
years=[2018,2019,2020,2021,2022,2023]
population=[10,20,30,40,50,60]
plt.scatter(years,population,s=salary,c='r',edgecolor='black' ,alpha=0.5)     ## alpha is to set the opacity
plt.show()
```

```python
salary[0]=salary[4]
plt.scatter(years,population,s=salary,c=salary,edgecolor='black')  
plt.title("Bubble Chart")
plt.show()
```

```python
## Adding text for each point
plt.scatter(years,population,s=salary,c=salary,edgecolor='black')  
plt.title("Bubble Chart")
for i in range(len(years)):
    plt.text(years[i]+0.2,population[i]+0.1,salary[i])
plt.show()
```

## pie chart

```python
import matplotlib.pyplot as plt
x=[6,4,9,5,2,15]
colors=['pink','blue','yellow','green','orange','brown']
labels=['A','B','C','D','E','F']
explode=[0,0,0,0,0,0.2]                    ## explode is used to separate parts from chart
plt.pie(x,colors=colors,explode=explode,labels=labels,autopct='%.1f%%',counterclock=False,startangle=100)  ##autopct will calc.its relative freq.distribution
plt.axis('equal')
plt.title('Pie chart ')
plt.show()
```

## Assignment Question:- Laptop sales

```python
import matplotlib.pyplot as plt
def laptopSales(sold,company):
    sum1=sum(sold)
    li=[]
    for i in range(len(sold)):
        num=(sold[i]*100)/sum1
        print(company[i],round(num,1))
        li.append(num)
    s=str(li)
    return s
sold=[20000,43000,15000,17000,22000,13000]
company=['HP','Dell','Lenovo','Apple','Acer','Asus']
print(laptopSales(sold,company))

plt.pie(sold,labels=company,counterclock=False,startangle=120,autopct='%.1f%%')
plt.show()
```

## Histogram 

```python
import numpy as np
import matplotlib.pyplot as plt
xt=np.arange(24)
x=[0,1,2,2,1,1,3,4,5,5,6,7,8,9,9,13,12,14,15,17,17,18,20,21,21,22,23,15]  ## in histogram formula for single line clubbed with no. is=max.no-min.no/10
plt.hist(x,edgecolor='black',bins=23)                ## we can specify bins for our accurate graph points , bins =given noi.e max-min/10
plt.xticks(xt)                               ## xticks represent all given points for x-axis plane
                                             ## how frequency distribution take place [a,b]=a  ,,, like for 1-2 =the no. of ones is three
plt.show()                                   ## but in last bin for graph line frequency both [a,b] are included like for 22-23 =freq is two                             
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_obj=pd.read_csv('amazon_jobs_dataset.csv')
file_obj.head()
file_obj.loc[2]
```

```python
file_obj['Year'] = pd.to_datetime(file_obj['Posting_date']).dt.year
```

```python
file_obj.loc[2]
```

## 1st . Assignment Qs: Line Graph for job per year

```python
posting_per_year=file_obj.groupby('Year').size()
per_year=file_obj['Year'].value_counts()
print(per_year)
print(posting_per_year)
plt.plot(posting_per_year.index, posting_per_year.values)
```

## 2nd assignment Qs : Bar graph for month wise job posting

```python
file_obj['Year'] = pd.to_datetime(file_obj['Posting_date']).dt.year
print(file_obj.loc[2])
file_obj['month']=pd.to_datetime(file_obj['Posting_date']).dt.month
print(file_obj.loc[2])
```

```python
import numpy as np
import matplotlib.pyplot as plt
job_posting_per_month=file_obj.groupby('month').size()
print(job_posting_per_month)
jp_posting=file_obj['month'].value_counts()
print(jp_posting)
##months=np.arange(13)
##job_posting_per_month = job_posting_per_month.reindex(months, fill_value=0)
plt.bar(job_posting_per_month.index,job_posting_per_month.values)
plt.show()
```

## 3rd Assignment Qs :pie chart for job opening cities in IN 

```python
file_obj.head()
```

```python
file_objIN = file_obj[file_obj['location'].str.contains('IN')]
#print(file_objIN)
#print(file_objIN.head())
file_objIN['City'] = file_objIN['location'].apply(lambda x: x.split(',')[-1].strip())
#.apply(): We use the apply() function to apply a custom function to each row in the location column.
#.split(','): The split(',') method splits the string by commas, turning it into a list of parts.
#[-1]: We select the last element of the list, which represents the city.
#.strip(): The strip() function removes any extra spaces around the city name.

job_openings_per_city=file_objIN['City'].value_counts()
print(job_openings_per_city)
plt.pie(job_openings_per_city,counterclock=False,labels=job_openings_per_city.index,startangle=120,autopct='%.2f%%')
```
