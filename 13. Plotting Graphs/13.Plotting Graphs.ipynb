## Matplotlib for data visualization

```python
import matplotlib.pyplot as plt
x=[1,4,65,5]
y=[4,64,23,100]
plt.scatter(x,y)           ## plt.scatter - will create a graph with dot points
plt.show()          
```

```python
import matplotlib.pyplot as plt
x=[1,4,65,5]
y=[4,64,23,100]
plt.plot(x,y)           ## plt.plot - will create a graph with line connecting dot points
plt.show()
```
